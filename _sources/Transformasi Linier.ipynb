# **Transformasi Linier**


## Definisi

Tranformasi adalah merepresentasikan suatu objek dari ruang dimensi dua ke dimensi tiga atau objek yang lain, atau memetakan ruang satu ke ruang yang lain.

### Cara menentukan tranformasi linier ada dua aturan:

- pusat titik tidak berubah
- garis lurus ditetapkan sebagai lurus

### Secara umum pemetaan $T$ adalah transformasi Linier jika

- $T(u+v)=T(u)+T(v )$
- $T(cu)=cT(u )$
 

## pembuktian bahwa sifat
$\displaystyle T(\mathbf{u}+\mathbf{v}) = (\mathbf{u})+T(\mathbf{v}$)

adalah

$T
\begin{pmatrix}
 \begin{bmatrix}
  u_1\\
  u_2\\
  u_3
 \end{bmatrix} +
 \begin{bmatrix}
  v_1\\
  v_2\\
  v_3
 \end{bmatrix}
\end{pmatrix}
$


- Misalkan kita memiliki matriks transformasi \($A$\) dan vektor-vektor \($\mathbf{u}$\) dan \($\mathbf{v}$\):

\[
$A = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix}, \quad \mathbf{u} = \begin{pmatrix}
x_1 \\
y_1
\end{pmatrix}, \quad \mathbf{v} = \begin{pmatrix}
x_2 \\
y_2
\end{pmatrix}$
\]


- Transformasi \($T$\) pada vektor \($\mathbf{u}$\) dan \($\mathbf{v}$\) adalah:

\[
$T(\mathbf{u}) = A\mathbf{u} = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix} \begin{pmatrix}
x_1 \\
y_1
\end{pmatrix} = \begin{pmatrix}
ax_1 + by_1 \\
cx_1 + dy_1
\end{pmatrix}$
\]

\[
$T(\mathbf{v}) = A\mathbf{v} = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix} \begin{pmatrix}
x_2 \\
y_2
\end{pmatrix} = \begin{pmatrix}
ax_2 + by_2 \\
cx_2 + dy_2
\end{pmatrix}$
\]

- hitung \( $T(\mathbf{u} + \mathbf{v})$ \):

\[
$\mathbf{u} + \mathbf{v} = \begin{pmatrix}
x_1 \\
y_1
\end{pmatrix} + \begin{pmatrix}
x_2 \\
y_2
\end{pmatrix} = \begin{pmatrix}
x_1 + x_2 \\
y_1 + y_2
\end{pmatrix}$
\]

- Kemudian, hitung \( $T(\mathbf{u} + \mathbf{v})$ \):

\[
$T(\mathbf{u} + \mathbf{v}) = A(\mathbf{u} + \mathbf{v}) = \begin{pmatrix}
a & b \\
c & d
\end{pmatrix} \begin{pmatrix}
x_1 + x_2 \\
y_1 + y_2
\end{pmatrix} = \begin{pmatrix}
a(x_1 + x_2) + b(y_1 + y_2) \\
c(x_1 + x_2) + d(y_1 + y_2)
\end{pmatrix}$
\]


- Sekarang, urai hasil di atas:

\[
$T(\mathbf{u} + \mathbf{v}) = \begin{pmatrix}
ax_1 + ax_2 + by_1 + by_2 \\
cx_1 + cx_2 + dy_1 + dy_2
\end{pmatrix}$
\]

Ini dapat ditulis ulang sebagai:

\[
$T(\mathbf{u} + \mathbf{v}) = \begin{pmatrix}
ax_1 + by_1 + ax_2 + by_2 \\
cx_1 + dy_1 + cx_2 + dy_2
\end{pmatrix} = \begin{pmatrix}
ax_1 + by_1 \\
cx_1 + dy_1
\end{pmatrix} + \begin{pmatrix}
ax_2 + by_2 \\
cx_2 + dy_2
\end{pmatrix}$
\]


Dengan ini, kita telah membuktikan bahwa untuk sebuah transformasi linier \($T$\) dan untuk semua vektor \($\mathbf{u}$\) dan \($\mathbf{v}$\), berlaku bahwa:

\[
$T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})$
\]



## pembuktian bahwa sifat

$\displaystyle T(c\mathbf{u}) = cT(\mathbf{u}$)

adalah

- Misalkan memiliki matriks transformasi \($ A $\) dan vektor \( $\mathbf{u}$ \):

\[
$A = \begin{pmatrix}
2 & 1 \\
0 & 3
\end{pmatrix}, \quad \mathbf{u} = \begin{pmatrix}
x \\
y
\end{pmatrix}$
\]

- Transformasi \( $T$ \) pada vektor \( $\mathbf{u}$ \) adalah:

\[
$T(\mathbf{u}) = A\mathbf{u} = \begin{pmatrix}
2 & 1 \\
0 & 3
\end{pmatrix} \begin{pmatrix}
x \\
y
\end{pmatrix} = \begin{pmatrix}
2x + y \\
3y
\end{pmatrix}$
\]


- misalkan \( $c$ \) adalah skalar. Transformasi \( $T$ \) pada vektor \( $c\mathbf{u}$ \) adalah:

\[
$c\mathbf{u} = c \begin{pmatrix}
x \\
y
\end{pmatrix} = \begin{pmatrix}
cx \\
cy
\end{pmatrix}$
\]

- Kemudian, hitung \( $T(c\mathbf{u})$ \):

\[
$T(c\mathbf{u}) = A(c\mathbf{u}) = \begin{pmatrix}
2 & 1 \\
0 & 3
\end{pmatrix} \begin{pmatrix}
cx \\
cy
\end{pmatrix} = \begin{pmatrix}
2(cx) + (cy) \\
3(cy)
\end{pmatrix} = \begin{pmatrix}
2cx + cy \\
3cy
\end{pmatrix}$
\]

- Sekarang, kalikan hasil transformasi \( $T(\mathbf{u})$ \) dengan skalar \( $c$ \):

\[
$cT(\mathbf{u}) = c \begin{pmatrix}
2x + y \\
3y
\end{pmatrix} = \begin{pmatrix}
c(2x + y) \\
c(3y)
\end{pmatrix} = \begin{pmatrix}
2cx + cy \\
3cy
\end{pmatrix}$
\]

- Membandingkan \( $T(c\mathbf{u})$ \) dan \( $cT(\mathbf{u})$ \)


\[
$T(c\mathbf{u}) = \begin{pmatrix}
2cx + cy \\
3cy
\end{pmatrix} = cT(\mathbf{u}) = \begin{pmatrix}
2cx + cy \\
3cy
\end{pmatrix}$
\]


Dengan ini, telah membuktikan bahwa untuk matriks transformasi \( $A$ \) dan vektor \( $\mathbf{u}$ \), serta skalar \( $c$ \):

\[
$T(c\mathbf{u}) = cT(\mathbf{u})$
\]



# **Tugas**

Transformasi \( $T$ \) pada vektor-vektor \($(2, 0)$\) dan \($(2, 2)$\) diberikan:

   \[
   $T(2, 0) = [-2, 0]$
   \]
   \[
   $T(2, 2) = [-2, 2]$
   \]

### Langkah-langkah Solusi:

#### 1. Basis Vektor dan Linear:

Untuk menemukan transformasi \( $T$ \), perlu menemukan \( $T$ \) pada basis vektor \( $\mathbf{e}_1$ \) dan \( $\mathbf{e}_2$ \):

\[
$\mathbf{e}_1 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \quad \mathbf{e}_2 = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$
\]

Kita dapat menuliskan vektor \($\begin{pmatrix} 2 \\ 0 \end{pmatrix}$\) dan \($\begin{pmatrix} 2 \\ 2 \end{pmatrix}$\) sebagai kombinasi linear dari basis vektor:

\[
$\begin{pmatrix} 2 \\ 0 \end{pmatrix} = 2 \mathbf{e}_1$
\]
\[
$\begin{pmatrix} 2 \\ 2 \end{pmatrix} = 2 \mathbf{e}_1 + 2 \mathbf{e}_2$
\]

#### 2. Menerapkan Transformasi \( $T$ \) pada Basis Vektor:

\[
$T(2\mathbf{e}_1) = 2T(\mathbf{e}_1)$
\]
\[
$T(2\mathbf{e}_1 + 2\mathbf{e}_2) = 2T(\mathbf{e}_1) + 2T(\mathbf{e}_2)$
\]

#### 3. Menggunakan Informasi tadi:

\( $T(2, 0) = [-2, 0]$\), kita bisa menuliskan:

\[
$T(2\mathbf{e}_1) = \begin{pmatrix} -2 \\ 0 \end{pmatrix}$
\]

Karena \( $T$ \) adalah linier, kita punya:

\[
$2T(\mathbf{e}_1) = \begin{pmatrix} -2 \\ 0 \end{pmatrix} \implies T(\mathbf{e}_1) = \begin{pmatrix} -1 \\ 0 \end{pmatrix}$
\]

Selanjutnya, diberikan \( $T(2, 2) = [-2, 2]$ \):

\[
$2T(\mathbf{e}_1) + 2T(\mathbf{e}_2) = \begin{pmatrix} -2 \\ 2 \end{pmatrix}$
\]

Lalu Substitusikan \( $T(\mathbf{e}_1)$ \):

\[
$2 \begin{pmatrix} -1 \\ 0 \end{pmatrix} + 2T(\mathbf{e}_2) = \begin{pmatrix} -2 \\ 2 \end{pmatrix}$
\]

\[
$\begin{pmatrix} -2 \\ 0 \end{pmatrix} + 2T(\mathbf{e}_2) = \begin{pmatrix} -2 \\ 2 \end{pmatrix}$
\]

\[
$2T(\mathbf{e}_2) = \begin{pmatrix} 0 \\ 2 \end{pmatrix} \implies T(\mathbf{e}_2) = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$
\]

#### 4. Membentuk Matriks Transformasi \( $A$ \):

Dengan \( $T(\mathbf{e}_1)$ \) dan \( $T(\mathbf{e}_2$) \), kita bisa bentuk matriks \( $A$ \):

\[
$A = \begin{pmatrix} T(\mathbf{e}_1) & T(\mathbf{e}_2) \end{pmatrix} = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix}$
\]

### Hasil:

Transformasi \( $T$ \) pada basis vektor \( $\mathbf{e}_1$ \) dan \( $\mathbf{e}_2$ \) ditemukan dan matriks \( $A$ \) yang merepresentasikan transformasi linier \( $T$ \) adalah:

\[
$A = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix}$
\]

# **Matriks Refleksi**

## Definisi
Refleksi atau pencerminan dalam konteks transformasi linier merujuk pada suatu transformasi geometri yang "mencerminkan" setiap titik dari suatu objek terhadap suatu garis (di 2D) atau suatu bidang (di 3D), menghasilkan gambar cerminnya. Transformasi ini adalah contoh dari transformasi linier yang dapat direpresentasikan menggunakan matriks.

### Refleksi di 2D
Refleksi di bidang dua dimensi dapat dilakukan terhadap sumbu x, sumbu y, garis y = x, atau garis y = -x. Matriks transformasi untuk refleksi tersebut adalah sebagai berikut:

**1. Refleksi terhadap sumbu x**:

   \[
   $\begin{pmatrix}
   1 & 0 \\
   0 & -1
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y) menjadi (x, -y).

**2. Refleksi terhadap sumbu y**:

   \[
   $\begin{pmatrix}
   -1 & 0 \\
   0 & 1
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y) menjadi (-x, y).

**3. Refleksi terhadap garis y = x**:

   \[
   $\begin{pmatrix}
   0 & 1 \\
   1 & 0
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y) menjadi (y, x).

**4. Refleksi terhadap garis y = -x**:

   \[
   $\begin{pmatrix}
   0 & -1 \\
   -1 & 0
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y) menjadi (-y, -x).

### Refleksi di 3D
Refleksi di ruang tiga dimensi dilakukan terhadap salah satu bidang koordinat (xy, xz, atau yz). Ada beberapa matriks transformasi untuk refleksi tersebut sebagai berikut:

**1. Refleksi terhadap bidang xy**:

   \[
   $\begin{pmatrix}
   1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & -1
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y, z) menjadi (x, y, -z).

**2. Refleksi terhadap bidang xz**:

   \[
   $\begin{pmatrix}
   1 & 0 & 0 \\
   0 & -1 & 0 \\
   0 & 0 & 1
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y, z) menjadi (x, -y, z).

**3. Refleksi terhadap bidang yz**:

   \[
   $\begin{pmatrix}
   -1 & 0 & 0 \\
   0 & 1 & 0 \\
   0 & 0 & 1
   \end{pmatrix}$
   \]
   Mengubah koordinat (x, y, z) menjadi (-x, y, z).
